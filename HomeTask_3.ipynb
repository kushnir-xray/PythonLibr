# Задание 1

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

boston = load_boston()
boston.keys()
```

```python
# Создайте датафреймы X и y из этих данных.
data = boston["data"]
data.shape
```

```python
feature_names = boston["feature_names"]
print(feature_names)
target = boston["target"]
target[:10]
```

```python
X = pd.DataFrame(data, columns=feature_names)
X.head()
```

```python
y = pd.DataFrame(target, columns=["value"])
y.info()
```

```python
# Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Создайте модель линейной регрессии под названием lr.
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Обучите модель на тренировочных данных (используйте все признаки)и сделайте предсказание. 
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(y_pred.shape)
check_test = pd.DataFrame({
 "y_test": y_test["price"],
 "y_pred": y_pred.flatten(),})
check_test.head(10)
```

```python
#Вычислите R2 полученных предказаний.
check_test["error"] = check_test["y_pred"] - check_test["y_test"]
print(check_test.head())
from sklearn.metrics import r2_score
r2_score_1=r2_score(check_test["y_pred"], check_test["y_test"])
r2_score_1
```

# Задание 2

```python
# Создайте модель под названием model с помощью RandomForestRegressor. 
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
# Обучите модель на тренировочных данных.
model.fit(X_train, y_train.values[:, 0])
```

```python
# Сделайте предсказание на тестовых данных и посчитайте R2.
y_pred = model.predict(X_test)
print(y_pred.shape)
check_test = pd.DataFrame({
 "y_test": y_test["price"],
 "y_pred": y_pred.flatten(),
})
print(check_test.head(10))
r2_score_2=r2_score(check_test["y_pred"], check_test["y_test"])
r2_score_2
```

```python
#Сравните с результатом из предыдущего задания.
r2_score_1<r2_score_2
```

```python
# RandomForest даёт лучшие результаты
```

# Задание 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
# Загрузите датасет creditcard.csv и создайте датафрейм df.
df = pd.read_csv('creditcard (3).csv')
df.head()
```

```python
# С помощью метода value_counts убедитесь в том, что выборка несбалансирована.
df["Class"].value_counts(normalize=True)
```

```python
# Используя метод info, проверьте, все ли столбцы содержат числовые данные
df.info()
```

```python
# Проверьте, нет ли в них пропусков.
df.isnull().astype(np.int).sum().astype(np.int)
```

```python
# Примените настройку, чтобы просматривать все столбцы датафрейма. 
pd.options.display.max_columns = 100
```

```python
# Просмотрите первые 10 строк датафрейма df.
df.head(10)
```

```python
# Создайте датафрейм X из датафрейма df, исключив столбец Class.
X = df.drop("Class", axis=1)
X.head()
```

```python
# Создайте объект Series под названием y из столбца Class.
y = df["Class"]
y.head()
```

```python
# Разбейте X и y на тренировочный и тестовый наборы данных.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y) 
```

```python
# Просмотрите информацию о их форме.
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Задайте параметры:
parameters = [{'n_estimators': [10, 15], 
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
# Создайте модель GridSearchCV.
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3)
```

```python
# Обучите модель на тренировочном наборе данных.
clf.fit(X_train, y_train)
```

```python
# Просмотрите параметры лучшей модели с помощью атрибута best_params_.
clf.best_params_
```

```python
# Предскажите вероятности классов с помощью полученнной модели.
y_pred_proba = clf.predict_proba(X_test)
print(y_pred_proba[:10])
```

```python
#Из полученного результата (массив Numpy) выберите столбец с индексом 1.
y_pred_proba = y_pred_proba[:, 1]
print(y_pred_proba[:5])
```

```python
# Из модуля sklearn.metrics импортируйте метрику roc_auc_score.
from sklearn.metrics import roc_auc_score
```

```python
# Вычислите AUC на тестовых данных и сравните с результатом тренировочных данных. 
roc_auc_score(y_test, y_pred_proba)
```
